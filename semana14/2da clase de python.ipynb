```python
def fx2(x,y):
    z=x*np.cos(x/y)-3*x**3
    return z
```

```python
def recta(a,m=1,b=0,label="x"):
    if m<0:
        print("Pendiente negativa encontrada. Cambie por m=0")
        m=0
        
    z=m*a+b
    plt.plot(a,z,label=label)
    return 
```
